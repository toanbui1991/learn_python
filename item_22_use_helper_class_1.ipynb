```python
class SimpleGradeBookd(object):
    """this class will each student grades by name using a dictionary

    Args:
        object (_type_): _description_
    """

    def __init__(self):
        self._grades = {}

    def add_student(self, name):
        self._grades[name] = []

    def report_grade(self, name, score):
        self._grades[name].append(score)

    def average_grade(self, name):
        grades = self._grades[name]
        return sum(grades) / len(grades)

class_book = SimpleGradeBookd()
class_book.add_student('toanbui1991')
class_book.add_student('buixuantoan1991')
class_book.report_grade('toanbui1991', 90)
class_book.report_grade('toanbui1991', 80)
class_book.report_grade('buixuantoan1991', 85)
class_book.report_grade('buixuantoan1991', 92)
toan_avg_grade = class_book.average_grade('toanbui1991')
bui_avg_grade = class_book.average_grade('buixuantoan1991')
print('toan_avg_grade: ', toan_avg_grade)
print('bui_avg_grade: ', bui_avg_grade)
```

```python
class BySubjectGradebook(object):
    def __init__(self):
        self._grades = {}


    def add_student(self, name):
        self._grades[name] = {}


    def report_grade(self, name, subject, grade):
        by_subject = self._grades[name]
        grade_list = by_subject.setdefault(subject, [])
        grade_list.append(grade)

    def average_grade(self, name):
        by_subject = self._grades[name]
        total, count = 0, 0
        for grades in by_subject.values():
            total += sum(grades)
            count += len(grades)
        return total / count

book = BySubjectGradebook()
book.add_student('Albert Einstein')
book.report_grade('Albert Einstein', 'Math', 75)
book.report_grade('Albert Einstein', 'Math', 65)
book.report_grade('Albert Einstein', 'Gym', 90)
book.report_grade('Albert Einstein', 'Gym', 95)
eistein_avg = book.average_grade("Albert Einstein")
print('eistein_avg: ', eistein_avg)
```

```python
from collections import namedtuple
Grade = namedtuple('Grade', ['score', 'weight'])

class Subject(object):

    def __init__(self, name):
        """initialize Subject with subject name

        Args:
            name (str): name of Subject
        """
        self.name = name
        self._grades = []

    def report_grade(self, score, weight):
        """add grade instance in self._grades

        Args:
            score (float): score
            weight (float): score
        """
        self._grades.append(Grade(score, weight))

    def average_score(self):
        """compute average score of subject

        Returns:
            float: average score of a given subject
        """
        total = 0
        total_weight = 0
        for grade in self._grades:
            total += grade.score * grade.weight
            total_weight += grade.weight
            avg = total / total_weight
        return avg

class Student(object):

    def __init__(self, name):
        self.name = name
        self._subjects = {}

    def subject(self, subject):
        """get subject given subject name

        Args:
            subject (str): subject

        Returns:
            Subject: subject instance of Subject    
        """
        if subject not in self._subjects:
            self._subjects[subject] = Subject(subject)
        return self._subjects[subject]
    
    def average_score(self):
        """compute avreage score of the student

        Returns:
            float: average score of student
        """
        total = 0
        count = 0
        for subject in self._subjects.values():
            total += subject.average_score()
            count += 1
        avg = total / count
        return avg

class GradeBook(object):

    def __init__(self, class_name):
        """initial instance of GradeBook with gradebook name

        Args:
            class_name (str): class_name of the gradebook
        """ 
        self.class_name = class_name
        self._students = {}

    def student(self, student):
        """get student instance from gradebook instance

        Args:
            student (str): student name

        Returns:
            Student: student instance of Student from gradebook
        """
        if student not in self._students:
            self._students[student] = Student(student)
        return self._students.get(student)


book = GradeBook('newName')
albert = book.student('Albert Einstein')
math = albert.subject('Math')
math.report_grade(80, 0.10)
math.report_grade(75, 0.90)
Gym = albert.subject('Gym')
Gym.report_grade(90, 0.10)
Gym.report_grade(95, 0.90)
print('abert average score: ', albert.average_score())

```
