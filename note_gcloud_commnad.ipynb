```python
#note about how to create compute engine in google cloud with gcloud command:

#one, compute engine name, define your own name
$VM_NAME='some_name'
#two, regions and zones, get zones list using command
gcloud compute zones list #chose asia-southeast1-c
#three, machine-types, get machine-types list with command
gcloud compute machine-types list --zones $ZONE_NAME #choose n1-standard-1
#to create compute engine (virtual machine or vm) in google cloud use command
gcloud compute instances create $VM_NAME --zone $ZONE_NAME - machine-type $VM_TYPE
#four, boot disk: is the image which we will use to install in the vm
    #with this config we have parameter (flags) like: --image, --image-project, --boot-disk-size
#five, service account and scope: this vm will use service account to call google api
    #scope: is assignment of what service account can do with google cloud api
    #so, vm is who the user, service account is the key to google cloud and scope is assign to that service account
    #what kind of action associate with the service account.
#six, extenal ip: we need extenal ip because we want to access this vm from the internet
    #use flag: --address to assign external ip to it
    #create external ip address: VPC Network > External IP addresses
    #note, external ip address will be created by default if not specify

#delete vm instance with command:
gcloud compute instances delete $VN_NAME --zone $ZONE_NAME
#note about compute engine (vm) name: name can not have underscore (_)
#note about service account: we have to create service account first, it is not create along with vm
#note about load environement varialbe from .env file with command: source .env

#put all together with final command:
gcloud compute instances create $VM_NAME --zone=$ZONE_NAME \
--machine-type=$VM_TYPE \
--image=$IMAGE \
--image-project=centos-cloud \
--boot-disk-size="50GB" \
--service-account=$SERVICE_ACCOUNT \
--scopes=$SCOPES

#note about reference link: https://medium.com/@kimeunsil/create-gce-vm-instance-with-gcloud-command-9bf20829090
```

```python
#Note about deploy jenkins in compute engine (vm)

#create vm
gcloud compute instances create $VM_NAME \
    --zone=$ZONE_NAME \
    --machine-type=$VM_TYPE \
    --subnet=default \
    --network-tier=PREMIUM \
    --maintenance-policy=MIGRATE \
    --tags=http-server,https-server \
    --image=ubuntu-1804-bionic-v20190918 \
    --image-project=ubuntu-os-cloud \
    --boot-disk-size=20GB \
    --boot-disk-type=pd-standard \
    --boot-disk-device-name=instance-1 \
    --reservation-affinity=any

#create firewall
#rules:tcp:8080, means allow internet traffic with tcp (Transmission Control Protocal)
#source-ranges, to filter api range can access mv
gcloud compute firewall-rules create $FIREWALL_NAME \
    --direction=INGRESS \
    --priority=1000 \
    --network=default \
    --action=ALLOW \
    --rules=tcp:8080 \
    --target-tags=http-server,https-server

#after create vm and firewall rule, we need to open the firewall and set up jenkins.
#link for reference:
    # https://faun.pub/jenkins-on-google-compute-engine-611bd86e295b
    # https://www.digitalocean.com/community/tutorials/how-to-install-jenkins-on-ubuntu-20-04

#note about delete

#to list all running compute engine instance with command
gcloud compute instances list --zone $ZONE_NAME
```
