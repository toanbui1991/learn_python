```python
#file.write vs file.writelines, file.read vs file.readlines
import numpy as np
number_generator = (str(np.random.randint(100)) + '\n' for _ in range(10000))
with open('large_file.txt', 'w') as file:
        file.writelines(number_generator)
```

```python
data = (x for x in open('./large_file.txt', 'r')) #this is generator
for d in data:
    print(int(d))
```
