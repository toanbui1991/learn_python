#### item38: Accept function instead of class for simmple interface
- how to use function as an input
- example is sorted(value, key) and defaultdict
- reference of sorted which supply custom key function: https://learnpython.com/blog/python-custom-sort-function/
- example with defaultdict(func, values). we have two example: 1 print add key and return 0, 2 is count the number of all missing keys
- exammple of supply defaultdict(instnace_of_counter, dict)

**Summary on use supply function for simple interface:**
- we can supply function as input for sorted(values, key)
- or define what to do if missing key in case of defaultdict(func, values)
- the other approach is supply instance of class for defaultdict(instance, values). this instance have __call__ method which execute when key is missing

```python
from collections import defaultdict
# Example 1, using function as in input
#reference sort with custom function: https://learnpython.com/blog/python-custom-sort-function/
names = ['Socrates', 'Archimedes', 'Plato', 'Aristotle']
names.sort(key=len)
print(names)

# Example 2, supply defaultdict(func, dict) with function.
#with supply functon if key is missing print key added and return 0 for that key
def log_missing():
    print('Key added')
    return 0

current = {'green': 12, 'blue': 3}
increments = [
    ('red', 5),
    ('blue', 17),
    ('orange', 9),
]
result = defaultdict(log_missing, current)
print('Before:', dict(result))
for key, amount in increments:
    result[key] += amount
print('After: ', dict(result))


#Example 3: count the number of missing key for all keys, 
def increment_with_report(current, increments):
    added_count = 0

    def missing():
        nonlocal added_count  # Stateful closure, nonlocal for function inside function
        added_count += 1
        return 0

    result = defaultdict(missing, current) #supply function as input
    for key, amount in increments:
        result[key] += amount

    return result, added_count



result, count = increment_with_report(current, increments)
assert count == 2
print(result)

#Example 4: 
class CountMissing:
    def __init__(self):
        self.added = 0

    def missing(self):
        self.added += 1
        return 0

counter = CountMissing()
result = defaultdict(counter.missing, current)  # Method ref
for key, amount in increments:
    result[key] += amount
assert counter.added == 2
print(result)

#Example 4, count missing key which input instance of a class and define __call__ which is called each if key is missing
class BetterCountMissing:
    def __init__(self):
        self.added = 0

    def __call__(self):
        self.added += 1
        return 0

counter = BetterCountMissing()
assert counter() == 0
assert callable(counter)

counter = BetterCountMissing()
result = defaultdict(counter, current)  # Relies on __call__
for key, amount in increments:
    result[key] += amount
assert counter.added == 2
print(result)
```
